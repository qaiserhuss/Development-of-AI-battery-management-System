```python
# 전체 데이터셋 가져오기
import pandas as pd
import os

# 로드할 디렉토리 경로 설정
load_directory = './DataForGPR/'

# 알려진 그룹명 리스트
selected_groups = ['2017-05-12', '2017-06-30', '2018-04-12']

# 로드한 데이터를 저장할 딕셔너리
loaded_data = {}
loaded_targets = {}

# 각 그룹별 데이터와 타겟 데이터 로드
for group_name in selected_groups:
    data_filename = os.path.join(load_directory, f"{group_name}_data.csv")
    target_filename = os.path.join(load_directory, f"{group_name}_target.csv")

    # 데이터 로드
    loaded_data[group_name] = pd.read_csv(data_filename)
    loaded_targets[group_name] = pd.read_csv(target_filename)

    print(f"Loaded {group_name} data from {data_filename}")
    print(f"Loaded {group_name} target from {target_filename}")

# 로드된 데이터 확인
for group_name in loaded_data.keys():
    print(f"Group {group_name} Data Shape: {loaded_data[group_name].shape}")
    print(f"Group {group_name} Target Shape: {loaded_targets[group_name].shape}")
```

```python
# 여기서부터 NAN 값과 아웃라이어 처리
def replace_nan_with_neighborhood_mean(data, neighborhood_size=5):
    # DataFrame을 NumPy 배열로 변환 (DataFrame인 경우)
    if isinstance(data, pd.DataFrame):
        data_np = data.to_numpy()
    else:
        data_np = data

    # 각 열에 대해 NaN 값 처리
    for col in range(data_np.shape[1]):
        nan_indices = np.where(np.isnan(data_np[:, col]))[0]
        for idx in nan_indices:
            start = max(0, idx - neighborhood_size)
            end = min(data_np.shape[0], idx + neighborhood_size + 1)
            mean_value = np.nanmean(data_np[start:end, col])
            data_np[idx, col] = mean_value

    return data_np if not isinstance(data, pd.DataFrame) else pd.DataFrame(data_np, columns=data.columns)

def replace_outliers_with_neighborhood_mean(data, neighborhood_size=5):
    # DataFrame을 NumPy 배열로 변환 (DataFrame인 경우)
    if isinstance(data, pd.DataFrame):
        data_np = data.to_numpy()
    else:
        data_np = data

    # 각 열에 대해 아웃라이어 처리
    for col in range(data_np.shape[1]):
        q1 = np.nanpercentile(data_np[:, col], 25)
        q3 = np.nanpercentile(data_np[:, col], 75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

        outlier_indices = np.where((data_np[:, col] < lower_bound) | (data_np[:, col] > upper_bound))[0]
        for idx in outlier_indices:
            start = max(0, idx - neighborhood_size)
            end = min(data_np.shape[0], idx + neighborhood_size + 1)
            mean_value = np.nanmean(data_np[start:end, col])
            data_np[idx, col] = mean_value

    return data_np if not isinstance(data, pd.DataFrame) else pd.DataFrame(data_np, columns=data.columns)
```

```python
# 각 그룹에 대해 NaN 값과 아웃라이어 처리
for group_name in selected_groups:
    # 데이터셋에서 NaN 값과 아웃라이어 처리
    loaded_data[group_name] = replace_nan_with_neighborhood_mean(loaded_data[group_name])
    loaded_data[group_name] = replace_outliers_with_neighborhood_mean(loaded_data[group_name])

    # 타겟 데이터셋에서 NaN 값과 아웃라이어 처리
    loaded_targets[group_name] = replace_nan_with_neighborhood_mean(loaded_targets[group_name])
    loaded_targets[group_name] = replace_outliers_with_neighborhood_mean(loaded_targets[group_name])
```

```python
# 처리된 데이터에 대해 정규화 적용
from sklearn.preprocessing import MinMaxScaler

# 각 그룹의 데이터와 타겟 데이터셋에 대한 MinMaxScaler 인스턴스 생성
scaler_data = MinMaxScaler()
scaler_target = MinMaxScaler()

# 모든 그룹의 데이터와 타겟 데이터에 스케일러 적용
for group_name in selected_groups:
    # 데이터 정규화
    loaded_data[group_name] = scaler_data.fit_transform(loaded_data[group_name])

    # 타겟 데이터 정규화
    loaded_targets[group_name] = scaler_target.fit_transform(loaded_targets[group_name])

# 정규화된 데이터 확인
for group_name in selected_groups:
    print(f"Group {group_name} Data After Scaling:\n", loaded_data[group_name])
    print(f"Group {group_name} Target After Scaling:\n", loaded_targets[group_name])
```

```python
import gpflow
import numpy as np
import tensorflow as tf

# 데이터셋 로드
# 가정: loaded_data와 loaded_targets에 각 그룹별 데이터가 저장되어 있음
X_train_1 = loaded_data['2017-05-12']
y_train_1 = loaded_targets['2017-05-12']
X_train_2 = loaded_data['2017-06-30']
y_train_2 = loaded_targets['2017-06-30']

# 훈련 데이터셋 결합
X_train = np.concatenate([X_train_1, X_train_2], axis=0)
y_train = np.concatenate([y_train_1, y_train_2], axis=0)

# 유도점의 개수 정의
num_inducing = 100  # 예시로 100개의 유도점 사용

# 커널 정의
kernel = gpflow.kernels.RBF()

# Sparse Gaussian Process 모델 생성
model = gpflow.models.SGPR(data=(X_train, y_train), kernel=kernel, inducing_variable=X_train[:num_inducing, :])

# 모델 최적화 (학습)
optimizer = gpflow.optimizers.Scipy()
optimizer.minimize(model.training_loss, variables=model.trainable_variables)
```

```python
import matplotlib.pyplot as plt

import gpflow
import numpy as np
import tensorflow as tf
from sklearn.metrics import mean_squared_error

# 테스트 데이터 그룹
test_group = '2018-04-12'

# 테스트 데이터셋 로드
X_test = loaded_data[test_group]
y_test = loaded_targets[test_group]

# 예측 수행
y_pred, var_pred = model.predict_f(X_test)

# 예측 결과 출력
print(f"Test Group: {test_group}")
print("Predictions:", y_pred.numpy().flatten())
print("Variances:", var_pred.numpy().flatten())

# RMSE 계산
rmse = np.sqrt(mean_squared_error(y_test, y_pred.numpy().flatten()))
print("Root Mean Squared Error:", rmse)

y_pred_transform = scaler_target.inverse_transform(y_pred)
y_test_transform = scaler_target.inverse_transform(y_test)

# for idx, i in enumerate(y_test_transform):
    # print("{} : {},  {}".format(idx, i, y_pred_transform[idx]))
    
# 실제 값과 예측 값의 시각화
plt.figure(figsize=(20, 6))
plt.plot(y_test_transform, label='Actual')
plt.plot(y_pred_transform, label='Predicted')
plt.title('Actual vs Predicted')
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```
